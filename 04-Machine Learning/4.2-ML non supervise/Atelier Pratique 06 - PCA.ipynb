##  Objectifs de cet atelier

Dans les projets de Machine Learning, nous travaillons souvent avec des données contenant un grand nombre de variables (features). Cependant, toutes ces variables ne sont pas toujours utiles et peuvent :

-  Compliquer l'analyse
-  Ralentir les modèles
-  Rendre la visualisation difficile

**La solution : PCA (Principal Component Analysis)**

Elle permet de :
-  Réduire le nombre de dimensions
-  Conserver un maximum d'information
-  Visualiser les données en 2D

### Principe du PCA


Imagine que tu as des données avec beaucoup de variables (dimensions) :

→ Exemple :

taille
poids
âge
revenu
etc...

→ Le problème :

• Trop de dimensions = difficile à visualiser

• Certaines variables sont corrélées (donc redondantes)

Peut-on simplifier les données sans perdre trop d’information ?

PCA cherche à :
→ La solution : PCA
Le PCA est une méthode qui permet de :

✔ réduire le nombre de variables

✔ garder le maximum d’information

L’idée clé du PCA

Au lieu d’utiliser les variables originales,
on va créer de nouvelles variables :

➡️ appelées composantes principales

C’est quoi une composante principale ?

 Une composante principale = une nouvelle variable construite à partir des anciennes

c’est un mélange des variables
ce n’est pas une variable réelle comme "âge" ou "taille"
