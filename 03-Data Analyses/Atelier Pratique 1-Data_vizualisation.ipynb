#   1. Définition de la Data Visualization


La Data Visualization (ou visualisation des données) est la discipline qui consiste à représenter des données sous forme de graphiques, de diagrammes ou de tableaux visuels.
La visualisation est une étape essentielle avant le Machine Learning car elle aide à mieux comprendre la structure et la nature du dataset.

Avec un graphique, le cerveau voit immédiatement :

*   la répartition
*   les valeurs extrêmes
*   voir si quelque chose augmente ou diminue dans le temps
*   les groupes


Notions fondamentales


*   Données : Informations brutes sous forme numérique ou textuelle.
*   Dataset : Ensemble structuré des données.

*   Feature (variable) : Colonne d'information dans un dataset.
*   Types de données :

    *   Numériques (quantitatives) : Age; Salaire ; Température …
    *   Catégoriques (qualitatives) : Sexe; Couleur; Type de logement …
    *   Temporelles : Historique des ventes

#4.2 Histogramme (distribution d’une variable)

```python
plt.hist(df["math score"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution des scores en Math")
plt.xlabel("Score")
plt.ylabel("Fréquence")
plt.show()
```

#4.3 Diagramme en barres (catégories)

```python
df["gender"].value_counts().plot(kind="bar", color=["pink", "blue"])
plt.title("Répartition par genre")
plt.xlabel("Genre")
plt.ylabel("Nombre d'étudiants")
plt.show()
```

#4.4 Line plot (courbe)

```python
df["math score"].plot(kind="line")
plt.title("Évolution des scores en Math")
plt.ylabel("Score")
plt.show()
```

# 4.5 Boxplot (identifier les valeurs extrêmes)

```python
plt.boxplot(df["math score"])
plt.title("Boxplot des scores en Math")
plt.ylabel("Score")
plt.show()
```

#4.6 Scatter plot (relation entre 2 variables)

```python
plt.scatter(df["reading score"], df["writing score"], color="purple")
plt.title("Lecture vs Écriture")
plt.xlabel("Reading score")
plt.ylabel("Writing score")
plt.show()
```

#Scatter plot 2 — Math score vs Average score

```python
df["average_score"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
```

```python
plt.scatter(df["math score"], df["average_score"], alpha=0.5, color="green")
plt.title("Score en math vs Score moyen")
plt.xlabel("Math score")
plt.ylabel("Average score")
plt.show()
```

```python
df.groupby("gender")["reading score"].mean().plot(kind="bar", color=["pink", "blue"])
plt.title("Score moyen en lecture selon le genre")
plt.ylabel("Score de lecture moyen")
plt.show()
```
